```python
import torch

data = torch.load('dataset_of_10_files.pt', weights_only=False)
data
print(data[0][0])
print(data[0][1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract first dataset (index 0)
first_dataset = data[0][0]

# Extract x and y values
x_values = [float(point[0]) for point in first_dataset]
y_values = [float(point[1]) for point in first_dataset]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, cmap='viridis', edgecolors='k', alpha=0.75)
plt.vlines(x_values, ymin=0, ymax=y_values, colors='gray', linestyles='dashed', alpha=0.6)
plt.xlabel("2Theta (degrees)")
plt.ylabel("Intensity")
plt.title("Scatter Plot of First XRD Dataset")

# Show plot
plt.show()
```
